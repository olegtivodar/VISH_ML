# Введение в NumPy

### Почему массивы NumPy в задачах обработки данных лучше базовой структуры Python - списков

1. Как в динамическом языке, все в python является объектом: помимо основной информации (которую мы обозначили явно) там хранятся дополнительные сведения: количество ссылок на объект, тип данных и т.д. Соответственно, если у нас огромные массивы данных, вся эта лишняя информация будет утяжелять и замедлять обработку
2. Операции над списками являются не самыми тривиальными - напрямую матричные операции с помощью встроенных операторов (+, -, \*) провести не получится - придется писать свои функции
3. В списках допускается использование различных типов данных одновременно, что не всегда хорошо: мы не можем сложить строку и число без приведения типов - если это не обработать явно, скрипты будут падать с ошибками

### Атрибуты массивов

Наиболее важные атрибуты объектов **ndarray**:
*   **ndarray.ndim** - число измерений (чаще их называют "оси") массива.
*   **ndarray.shape** - размеры массива, его форма. Это кортеж натуральных чисел, показывающий длину массива по каждой оси. Для матрицы из n строк и m столбов, shape будет (n,m). Число элементов кортежа shape равно ndim.
*   **ndarray.size** - количество элементов массива. Очевидно, равно произведению всех элементов атрибута shape.
*   **ndarray.dtype** - объект, описывающий тип элементов массива. Можно определить dtype, используя стандартные типы данных Python. NumPy здесь предоставляет целый букет возможностей, как встроенных, например: bool_, character, int8, int16, int32, int64, float8, float16, float32, float64, complex64, object_, так и возможность определить собственные типы данных, в том числе и составные.
*   **ndarray.itemsize** - размер каждого элемента массива в байтах.
*   **ndarray.data** - буфер, содержащий фактические элементы массива. Обычно не нужно использовать этот атрибут, так как обращаться к элементам массива проще всего с помощью индексов.

### Сводные показатели и агрегаты

```python
x = np.arange(1, 6)
x
```

```python
np.add.reduce(x)
```

```python
np.multiply.reduce(x) #multiply - произведение (по элементам)
```

```python
# сохранение промежуточного результата

np.add.accumulate(x) #add - сумма (по элементам) + accumulate
```

```python
np.multiply.accumulate(x)
```

```python
# Все универсальные функции могут выводить результат применения
# соответствующей операции ко всем парам двух аргументов
# с помощью метода outer

x = np.arange(1, 6)
```

```python
np.multiply.outer(x, x) #outer - Внешнее «произведение»
```

Т.е. Бинарные функции (функции двух аргументов) поддерживают дополнительные методы, позволяющие накапливать значения результата различными способами.

*   **accumulate()** - аккумулирует результат применения оператора ко всем элементам
*   **outer()** - Внешнее «произведение» - применение оператора к парам из двух массивов
*   **reduce()** - результат применения оператора ко всем элементам

### **Задача 7**
Посчитайте метрику accuracy для следующих входных данных:

y_true = ["positive", "negative", "negative", "positive", "positive", "positive", "negative"]

y_pred = ["positive", "negative", "positive", "positive", "negative", "positive", "positive"]


**План действий:**
1. Что такое accuracy? В решении каких задач используется эта метрика?
2. Как она считается? (подсказка: посмотрите что такое confusion matrix)

Помним, что:
positive = 1
negative = 0

```python
#Код решения
y_true = np.array([0 if i == "negative" else 1 for i in ["positive", "negative", "negative", "positive", "positive", "positive", "negative"]])
y_pred = np.array([0 if i == "negative" else 1 for i in ["positive", "negative", "positive", "positive", "negative", "positive", "positive"]])

correct_predictions = np.sum(y_true == y_pred)

total_predictions = len(y_true)

accuracy = correct_predictions / total_predictions

print(f"Accuracy: {accuracy}")
```

### **Задача 8**
Создайте 3 массива размерностями (3, 1), (3, 1) и (2, 1) соответственно.
Сделайте их последовательное слияние, а затем найдите для каждого элемента массива остаток от деления на 3. В получившемся массиве найдите наибольшее значение.

```python
#Код решения
arr1 = np.array([[5], [8], [11]])
arr2 = np.array([[2], [7], [10]])
arr3 = np.array([[3], [6]])

merged_array = np.vstack((arr1, arr2, arr3))

remainder_array = merged_array % 3

max_value = np.max(remainder_array)

print("Объединенный массив:")
print(merged_array)

print("Массив с остатками от деления на 3:")
print(remainder_array)

print("Наибольшее значение в массиве:", max_value)
```

Функция **numpy.dot(a, b, out=None)** вычисляет скалярное произведение двух массивов.

Перед использованием данной функции необходимо учитывать следующие нюансы:



*   Если a или b является **числом**, то np.dot(a, b) становится эквивалентна функции np.multiply(a, b) или команде a*b, которые являются более предпочтительными;
*   Если a или b являются **одномерными массивами**, т.е. векторами, то выполняется внутреннее (по сути скалярное) произведение векторов и np.dot(a, b) становится эквивалентна функции np.inner(a, b), но без поддержки комплексных чисел.
*   Если **a - это N-мерный массив**, **b - это одномерный массив**, то сумма произведений элементов a и b вычисляется по последней оси a.
*   Если** a и b являются двумерными массивами**, т.е. матрицами, то выполняется матричное умножение и np.dot(a, b) становится эквивалентна функции np.matmul(a, b) или команде a@b, которые являются более предпочтительными;
*   Если **a - это N-мерный массив**, а **b - это M-мерный массив (при условии, что M >= 2)**, то вычисляется сумма произведений элементов последней оси a и элементов со второй по последнюю осей b, т.е. эквивалентно команде: dot(a, b)[i,j,k,m] = sum(a[i,j,:] * b[k,:,m])

### **Задача 11**
Создайте случайный массив (5 на 5) и найдите наиболее часто встречающееся значение в этом массиве

```python
#Код решения

arr = np.random.randint(10, size = (5, 5))

flattened_arr = arr.flatten()
value_counts = np.bincount(flattened_arr)
most_common_value = np.argmax(value_counts)

print("Случайный массив:")
print(arr)
print(f"Наиболее часто встречающееся значение: {most_common_value}")
```

**np.flatten** - Преобразовывает массивы в одномерный вектор

**np.bincount** - Подсчитывает количество вхождений каждого значения в массиве неотрицательных целых чисел.

**np.ndarray.argmax** - Возвращает индексы максимальных значений по заданной оси.
